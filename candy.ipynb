```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Candy.csv")

df.head()
```

```python
print(df.shape)
print(df.columns)
df.info()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df[df['Sales'] <= 0]
```

```python
df = df[df['Sales'] > 0]
```

```python
df[df['Cost'] < 0]
```

```python
df = df[df['Cost'] >= 0]
```

```python
df['Units'].isnull().sum()
```

```python
df['Units'] = df['Units'].fillna(df['Units'].median())
```

```python
df['Product Name'] = df['Product Name'].str.strip()
```

```python
df['Division'] = df['Division'].str.strip().str.title()
```

```python
df.info()
```

```python
df['Gross Margin %'] = (
    df['Gross Profit'] / df['Sales']
) * 100
```

```python
df['Profit Per Unit'] = (
    df['Gross Profit'] / df['Units']
)
```

```python
product_summary = df.groupby(
    'Product Name'
).agg({
    'Sales':'sum',
    'Gross Profit':'sum',
    'Cost':'sum',
    'Units':'sum'
}).reset_index()

product_summary
```

```python
product_summary['Gross Margin %'] = (
    product_summary['Gross Profit']
    / product_summary['Sales']
) * 100
product_summary['Profit Per Unit'] = (
    product_summary['Gross Profit']
    / product_summary['Units']
)
total_sales = product_summary['Sales'].sum()

product_summary['Revenue Contribution %'] = (
    product_summary['Sales']
    / total_sales
) * 100
total_profit = product_summary['Gross Profit'].sum()

product_summary['Profit Contribution %'] = (
    product_summary['Gross Profit']
    / total_profit
) * 100
top_profit_products = product_summary.sort_values(
    by='Gross Profit',
    ascending=False
)

top_profit_products.head(10)
```

```python
top_margin_products = product_summary.sort_values(
    by='Gross Margin %',
    ascending=False
)

top_margin_products.head(10)
```

```python
division_summary = df.groupby(
    'Division'
).agg({
    'Sales':'sum',
    'Gross Profit':'sum',
    'Cost':'sum'
}).reset_index()

division_summary
```

```python
division_summary['Margin %'] = (
    division_summary['Gross Profit']
    / division_summary['Sales']
) * 100

division_summary
```

```python
pareto_df = product_summary.sort_values(
    by='Gross Profit',
    ascending=False
)
pareto_df['Cumulative Profit'] = (
    pareto_df['Gross Profit'].cumsum()
)
pareto_df['Cumulative Profit %'] = (
    pareto_df['Cumulative Profit']
    / total_profit
) * 100
pareto_df[
    pareto_df['Cumulative Profit %'] <= 80
]
```

```python
def risk_flag(margin):

    if margin < 20:
        return "High Risk"

    elif margin < 40:
        return "Medium Risk"

    else:
        return "Low Risk"
     
```

```python
   product_summary['Risk Flag'] = (
    product_summary['Gross Margin %']
    .apply(risk_flag)
)
```
